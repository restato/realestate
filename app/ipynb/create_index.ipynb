```python
#-*- coding: utf-8 -*-
from elasticsearch import Elasticsearch,JSONSerializer, helpers
import numpy as np
import json

# https://github.com/elastic/elasticsearch-py/issues/378
class NumpyEncoder(JSONSerializer):
    """ Special json encoder for numpy types """
    def default(self, obj):
        if isinstance(obj, (np.int_, np.intc, np.intp, np.int8,
            np.int16, np.int32, np.int64, np.uint8,
            np.uint16, np.uint32, np.uint64)):
            return int(obj)
        elif isinstance(obj, (np.float_, np.float16, np.float32, 
            np.float64)):
            return float(obj)
        elif isinstance(obj,(np.ndarray,)): #### This is the fix
            return obj.tolist()
        return JSONSerializer.default(self, obj)
        
es = Elasticsearch(hosts="122.32.196.201:9200", port=9200, serializer=NumpyEncoder())  # 객체 생성

# Es Cluster Health Check
es.cluster.health()
# es.indices.delete(index='apt-rent') 
es.indices.delete(index='apt-trade') 
# index 생성
# es.indices.create(index='apt-rent', body=apt_rent_body)
es.indices.create(index='apt-trade', body=apt_trade_body)
# index 목록
print(es.cat.indices())
```
